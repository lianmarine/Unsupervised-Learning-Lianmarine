```python
import requests
import csv
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
from yellowbrick.cluster import KElbowVisualizer
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
```

```python
dataset = pd.read_csv("https://raw.githubusercontent.com/4GeeksAcademy/k-means-project-tutorial/main/housing.csv")
df = pd.DataFrame(dataset)
df
```

```python
df_new = df[["MedInc", "Latitude", "Longitude"]]
df_new
```

```python
X_train, X_test = train_test_split(df_new, test_size=0.2, random_state=42)
```

```python
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

model = KMeans(n_clusters = 5, random_state = 42)
model.fit(X_train)
```

```python
X_train["Cluster"] = list(model.labels_)
y_train = X_train["Cluster"]
```

```python
X_train
```

```python
# Clase KElbowVisualizer para visualizar el codo
elbow_vis = KElbowVisualizer(KMeans(random_state=42), k=(1, 20))
elbow_vis.fit(X_train)
elbow_vis.show()
```

```python


# Assuming df is your DataFrame with the 'Cluster' column

fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot for the complete set
scatter = ax.scatter(X_train['Longitude'], X_train['Latitude'], X_train['MedInc'], c=X_train['Cluster'], cmap='viridis', s=10, label='Complete Set')

ax.set_title('K-Means Clustering (3D)')
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
ax.set_zlabel('MedInc')

# Set specific view angles
ax.view_init(elev=20, azim=30)  # Adjust the elevation (up/down) and azimuth (rotation around z-axis) angles

# Add a legend
legend = ax.legend(*scatter.legend_elements(), title='Clusters')
ax.add_artist(legend)

plt.show()
```

```python
y_test = list(model.predict(X_test))
X_test["Cluster"] = y_test
X_test.head()
```

In order to prove thge accuracy of the clustering, the values obtained by the moel on the test set are to be added to the plot. These are represented with the same colour as in the original plot, but by '+' signs instead of dots.

```python
# Assuming df is your DataFrame with the 'Cluster' column

fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot for the complete set
scatter1 = ax.scatter(X_train['Longitude'], X_train['Latitude'], X_train['MedInc'], c=X_train['Cluster'], cmap='viridis', s=10, label='Complete Set')

# Scatter plot for the second set with '+'
scatter2 = ax.scatter(X_test['Longitude'], X_test['Latitude'], X_test['MedInc'], c=X_test['Cluster'],cmap='viridis', marker='+', s=50, label='Second Set')

ax.set_title('K-Means Clustering (3D)')
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
ax.set_zlabel('MedInc')

# Set specific view angles
ax.view_init(elev=20, azim=30)  # Adjust the elevation (up/down) and azimuth (rotation around z-axis) angles

# Add legends for both sets
legend1 = ax.legend(*scatter1.legend_elements(), title='Clusters - Complete Set')
legend2 = ax.legend(*scatter2.legend_elements(), title='Second Set', loc="upper left")
ax.add_artist(legend1)

plt.show()
```

The clustring provided by the training set shows a great deal of accuracy when applied with the test set, the model can therefore be deemed as successful.


USE DECISION TREE ALOGORYTHM AS AN ALTERNATIVE CLASSIFICATION MODEL

```python
from sklearn.tree import DecisionTreeClassifier

model_sup = DecisionTreeClassifier(random_state = 42)
model_sup.fit(X_train, y_train)
```

```python
from sklearn import tree

fig = plt.figure(figsize=(15,15))

tree.plot_tree(model_sup, feature_names = list(X_train.columns), class_names = ["0", "1", "2", "3", "4", "5"], filled = True)

plt.show()
```

```python
y_pred = model_sup.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_pred)
```
